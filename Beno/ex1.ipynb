# Practicals 1 - Beta distribution

## PDF (a=0.5, b=0.5)

```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt


fig, ax = plt.subplots(1, 1, figsize=(10, 10))

a, b = 0.5, 0.5
x = np.linspace(beta.ppf(0.01, a, b),
                beta.ppf(0.99, a, b), 100)

ax.plot(x, beta.pdf(x, a, b),
       'r-', lw=5, alpha=0.6, label='beta pdf')

half = beta.ppf(0.5, a, b)
ax.axhline(y=half, c='blue')
```

## CDF

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
x = np.linspace(beta.ppf(0.01, a, b),
                beta.ppf(0.99, a, b), 100)

ax.plot(x, beta.cdf(x, a, b),
       'r-', lw=5, alpha=0.6, label='beta cdf')
```

```python
beta.cdf(x, a, b)
```
